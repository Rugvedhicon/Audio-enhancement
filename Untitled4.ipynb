```python
# Install required packages
!pip install tensorflow librosa==0.10.0 soundfile==0.12.1 matplotlib numpy --quiet

# Imports
import os
import warnings
import numpy as np
import librosa
import librosa.display
import soundfile as sf
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from IPython.display import Audio, display

warnings.filterwarnings('ignore')
np.random.seed(42)
tf.random.set_seed(42)
os.makedirs("sample_data", exist_ok=True)
```

```python

def download_sample_audio():
    audio_path = librosa.example('trumpet')
    y, sr = librosa.load(audio_path, sr=22050)
    sf.write('sample_data/high_quality.wav', y, sr)
    # Low-quality version (downsample then upsample)
    y_low = librosa.resample(y, orig_sr=sr, target_sr=8000)
    y_low = librosa.resample(y_low, orig_sr=8000, target_sr=sr)
    sf.write('sample_data/low_quality.wav', y_low, sr)
    return 'sample_data/high_quality.wav', 'sample_data/low_quality.wav'

high_quality_path, low_quality_path = download_sample_audio()
print(f"Data saved: {high_quality_path}, {low_quality_path}")
```

```python
def create_synthetic_dataset(high_quality_path, num_samples=100, duration=1.0):
    y_hq, sr = librosa.load(high_quality_path, sr=22050)
    segment_len = int(duration * sr)
    hq_segments, lq_segments = [], []
    
    for _ in range(num_samples):
        start = np.random.randint(0, len(y_hq) - segment_len)
        segment = y_hq[start:start + segment_len]
        
        # Low quality version
        segment_lq = librosa.resample(segment, orig_sr=sr, target_sr=8000)
        segment_lq = librosa.resample(segment_lq, orig_sr=8000, target_sr=sr)
        
        min_len = min(len(segment), len(segment_lq))
        hq_segments.append(segment[:min_len])
        lq_segments.append(segment_lq[:min_len])
    
    return np.array(hq_segments), np.array(lq_segments), sr

hq_audio, lq_audio, sr = create_synthetic_dataset(high_quality_path)
print(f"Dataset: {len(hq_audio)} samples, SR: {sr}")
```

```python
def audio_to_spectrogram(audio, n_fft=1024, hop_length=256):
    spec = librosa.stft(audio, n_fft=n_fft, hop_length=hop_length)
    return np.abs(spec), np.angle(spec)

def spectrogram_to_audio(magnitude, phase, hop_length=256):
    return librosa.istft(magnitude * np.exp(1j * phase), hop_length=hop_length)

def prepare_training_data(hq_audio, lq_audio, n_fft=1024, hop_length=256):
    X, y = [], []
    for lq, hq in zip(lq_audio, hq_audio):
        lq_mag, _ = audio_to_spectrogram(lq, n_fft, hop_length)
        hq_mag, _ = audio_to_spectrogram(hq, n_fft, hop_length)
        
        lq_mag = librosa.amplitude_to_db(lq_mag, ref=np.max)[..., np.newaxis]
        hq_mag = librosa.amplitude_to_db(hq_mag, ref=np.max)[..., np.newaxis]
        
        X.append(lq_mag)
        y.append(hq_mag)
    
    return np.array(X), np.array(y)

X_train, y_train = prepare_training_data(hq_audio, lq_audio)
print(f"Shapes: X={X_train.shape}, y={y_train.shape}")

```

```python
input_shape = X_train.shape[1:]  # (freq_bins, time_frames, 1)

def create_simple_unet(input_shape):
    inputs = layers.Input(shape=input_shape)
    
    # Encoder
    x1 = layers.Conv2D(16, (3,3), activation='relu', padding='same')(inputs)
    x1 = layers.Conv2D(16, (3,3), activation='relu', padding='same')(x1)
    p1 = layers.MaxPooling2D((2,2))(x1)
    
    x2 = layers.Conv2D(32, (3,3), activation='relu', padding='same')(p1)
    x2 = layers.Conv2D(32, (3,3), activation='relu', padding='same')(x2)
    p2 = layers.MaxPooling2D((2,2))(x2)
    
    # Bottleneck
    b = layers.Conv2D(64, (3,3), activation='relu', padding='same')(p2)
    b = layers.Conv2D(64, (3,3), activation='relu', padding='same')(b)
    
    # Decoder
    u2 = layers.UpSampling2D((2,2))(b)
    u2 = layers.Conv2D(32, (3,3), activation='relu', padding='same')(u2)
    u2 = layers.Concatenate()([u2, x2])
    
    u1 = layers.UpSampling2D((2,2))(u2)
    u1 = layers.Conv2D(16, (3,3), activation='relu', padding='same')(u1)
    u1 = layers.Concatenate()([u1, x1])
    
    outputs = layers.Conv2D(1, (1,1), activation='linear')(u1)
    return models.Model(inputs, outputs)

model = create_simple_unet(input_shape)
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()
```

```python
# Train the new model
history = model.fit(
    X_train,
    y_train,
    batch_size=8,
    epochs=20,
    validation_split=0.2,
    verbose=1
)
```

```python
import matplotlib.pyplot as plt

# Step 8: Training History Plot
plt.figure(figsize=(12, 4))

# Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()

# MAE
plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Train MAE')
plt.plot(history.history['val_mae'], label='Val MAE')
plt.title('MAE over Epochs')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def enhance_audio(lq_audio, model, n_fft=1024, hop_length=256):
    # Convert LQ audio to spectrogram
    lq_mag, phase = audio_to_spectrogram(lq_audio, n_fft, hop_length)
    lq_mag_db = librosa.amplitude_to_db(lq_mag, ref=np.max)

    # Prepare input for the model
    inp = lq_mag_db[np.newaxis, ..., np.newaxis]  # shape: (1, freq, time, 1)

    # Predict enhanced magnitude
    enhanced_db = model.predict(inp, verbose=0)[0, ..., 0]

    # Convert back to amplitude
    enhanced_mag = librosa.db_to_amplitude(enhanced_db)

    # Reconstruct audio with original phase
    enhanced_audio = spectrogram_to_audio(enhanced_mag, phase, hop_length)

    return enhanced_audio
```

```python
# Pick a sample from your dataset
test_idx = 0
lq_test, hq_test = lq_audio[test_idx], hq_audio[test_idx]

# Enhance it
enhanced_audio = enhance_audio(lq_test, model)

# Save audios to listen
sf.write('sample_data/lq_test.wav', lq_test, sr)
sf.write('sample_data/hq_test.wav', hq_test, sr)
sf.write('sample_data/enhanced.wav', enhanced_audio, sr)

# Listen in notebook
print("Low Quality:"); display(Audio('sample_data/lq_test.wav'))
print("High Quality:"); display(Audio('sample_data/hq_test.wav'))
print("Enhanced:"); display(Audio('sample_data/enhanced.wav'))
```

```python
import librosa.display

def plot_specs(lq, hq, enhanced, sr=22050, n_fft=1024, hop_length=256):
    plt.figure(figsize=(12, 8))

    for i, (audio, title) in enumerate(zip([lq, hq, enhanced],
                                           ["Low Quality", "High Quality", "Enhanced"])):
        plt.subplot(3, 1, i+1)
        mag, _ = audio_to_spectrogram(audio, n_fft=n_fft, hop_length=hop_length)
        db = librosa.amplitude_to_db(mag, ref=np.max)
        librosa.display.specshow(db, sr=sr, hop_length=hop_length, x_axis='time', y_axis='log')
        plt.title(f"{title} Spectrogram")
        plt.colorbar(format='%+2.0f dB')

    plt.tight_layout()
    plt.show()

# Plot for our test sample
plot_specs(lq_test, hq_test, enhanced_audio, sr)
```

```python
def test_custom_audio(path, model, target_sr=22050, n_fft=1024, hop_length=256):
    # Load custom audio with target sample rate
    y, orig_sr = librosa.load(path, sr=target_sr)
    
    # Create low-quality version (downsample then upsample)
    y_low = librosa.resample(y, orig_sr=target_sr, target_sr=8000)
    y_low = librosa.resample(y_low, orig_sr=8000, target_sr=target_sr)
    
    # Trim to same length
    min_len = min(len(y), len(y_low))
    y = y[:min_len]
    y_low = y_low[:min_len]
    
    # Enhance using the model
    enhanced = enhance_audio(y_low, model, n_fft=n_fft, hop_length=hop_length)
    
    # Save outputs
    base, ext = os.path.splitext(path)
    sf.write(base + '_low.wav', y_low, target_sr)
    sf.write(base + '_enhanced.wav', enhanced, target_sr)
    
    return y_low, enhanced

# Test with your custom audio
custom_path = "Audi.mp3"
lq_custom, enhanced_custom = test_custom_audio(custom_path, model)

# Display results
print("Low Quality Custom:"); display(Audio(lq_custom, rate=22050))
print("Enhanced Custom:"); display(Audio(enhanced_custom, rate=22050))
```
